```python
def loadDataSet(fileName):
    en_records=[]
    cn_records=[]
    records=[]
    fr=open(fileName,encoding="utf-8")
    for line in fr:
        if line.strip() :  
            records.append(line.strip())
        
    for index,line in enumerate(records):
        if index %2 ==0:
            cn_records.append(line)
        else:
            en_records.append(line)
    return cn_records,en_records
```

                           1.翻译不准确（因为专有名词，特殊字符，时态等造成）
                           
                           2.句法结构（主谓宾搭配不当，句子有逻辑错误，不止存在于个别词，而是整个句子）
                           
                           3.错误传播（某些词反复出现，不仅影响当前句子，更影响到后文，属于句子与句子之间）
                           
                           4.没有翻译（包括个别词语，以及句子级别的漏译）
                           
                           5.分词错误（中文分词错误，对分词工具的顺带评测）
                           
                           6.没有分句（将多个句子整合为一句子翻译，以致造成翻译错误，通常会引发其他类型错误，比如1，4，5）

# 仔细观察每个样本对应的错误类型，我们不难发现，第一类错误：翻译不准确是一个跨样本的错误类型，在每个样本中错误数量都名列前茅。因此解决此类问题使我们的首要目标，紧接着，第四类错误：没有翻译，在不同的样本中水平在起伏，说明它并不是一个广泛存在的问题，只是在个别样本，比如说A,B,D,F样本中情况比较严重，这个错误类型需要针对具体样本来进行具体分析。第六类错误：没有分句，数量在每个样本中都是居中水平，但是一直存在，说明这是一个广泛存在的问题，但并不是一个严重的问题，需要进行具体的样本再抽取量化分析。我们也可以关注一些个性的问题，比如说，B，D样本1，4类错误比较严重；C样本3类错误远高于其他样本。
